```python
from dotenv import load_dotenv
import os
load_dotenv('.env')
os.environ['LANGCHAIN_ENDPOINT']
```

```python
from langchain_ollama import ChatOllama

base_url = "http://localhost:11434" # 서버 주소
model = 'eeve:q4'                   # 한글 사용 모델 이름
# model = 'llama3.2:1b'             # 모델 이름

llm = ChatOllama(
    base_url=base_url,
    model = model,
    temperature = 0.8,
    num_predict = 256,
)

response = llm.invoke('대한민국에 대해 5문장으로 설명해줘')
print(response.content)
```

```python
# 모든 컨텐츠 생성과정 확인
response = ""
for chunk in llm.stream('대한민국에 대해 2문장으로 설명해줘'):
    print(chunk)
    response = response + " " + chunk.content
    print(response)
```
